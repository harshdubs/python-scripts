```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

plt.style.use('seaborn-v0_8')   #sets a clean visual style for all charts.
sns.set_palette("husl")

print("Libraries loaded successfully!")
```

```python
df = pd.read_csv('data/DataCOSupplyChainDataset.csv',encoding='latin 1')

df.head()
```

```python
df.columns.tolist()
```

```python
discounted = df[df['Order Item Discount Rate'] > 0]
non_discounted = df[df['Order Item Discount Rate'] == 0]

print(discounted['Benefit per order'].mean())
print(non_discounted['Benefit per order'].mean())
```

Hypothesis testing: 

H₀ - Null hypothesis - There is no difference in profit between orders with discount and without discount - True untill proven wrong by stats and data

H₁ - Aternative Hypothesis - there might be something, there might be a difference - My accusation

α  - threshold value, if p-value is less than this, then null hypothesis Ho is proven false = 0.05 = probablitiy lower than 5% that difference is random noise 


```python
from scipy.stats import ttest_ind 

t_val,p_val = ttest_ind(discounted['Benefit per order'], non_discounted['Benefit per order'], equal_var = False)

print(f'T value : {t_val}')
print(f'P Value : {p_val}')
```

```python
from math import sqrt

mean1 = discounted['Benefit per order'].mean()
mean2 = non_discounted['Benefit per order'].mean()
n1 = len(discounted)
n2 = len(non_discounted)
s1 = discounted['Benefit per order'].std()
s2 = non_discounted['Benefit per order'].std()

pooled_std = sqrt( ((n1-1)*s1**2 + (n2-1)*s2**2) / (n1 + n2 - 2) )

d = (mean1 - mean2) / pooled_std

print(f'Cohen d = {d.round(3)}')
```

We reject/accept H₀ because p = 0.0001 which is below α = 0.05.
Discounted orders averaged $21.6 profit vs $26.6 for non-discounted.
However, Cohen's d = -0.048 which indicates a insignificant effect size.
This means we can conclude that there is difference from null hypothesis, discount negatively impacts the benefit per order, but even though the p value is less, it just gaurentees that there is impact from discounts nut our d-value which is way lesser than 0.2 hence the impact is really low on benefit.
